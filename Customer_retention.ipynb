# univiriate Analysis

```python
df['age_group']=df[ '2 How old are you? ']
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='age_group',data=df)
plt.title('age_group count')
plt.show()


print(df['age_group'].value_counts())
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='gender',data=df)      
plt.title('gender count')
plt.show()

print("0 = Male")
print("1 = Female")
print(df['gender'].value_counts())
```

```python
Here female customer are more than male
```

```python
plt.subplots(figsize = (16,6))
sns.countplot(x='3 Which city do you shop online from?',data=df)      
plt.title('city ')
plt.show()


print(df['3 Which city do you shop online from?'].value_counts())
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='5 Since How Long You are Shopping Online ?',data=df)      
plt.title('Since How Long You are Shopping Online')
plt.show()

print('1 = less than 1 year')
print('2 = 1-2')
print('3 = 2-3')
print('4 = 3-4')
print('5 = more than 4 years')

print(df['5 Since How Long You are Shopping Online ?'].value_counts())
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='6 How many times you have made an online purchase in the past 1 year?',data=df)      
plt.title('no. of order made in past 1 year')
plt.show()

print('1 = less than 10 times')
print('2 = 11-20 times')
print('3 = 21-30 times')
print('4 = 31-40 times')
print('5 = 41 and above')

print(df['6 How many times you have made an online purchase in the past 1 year?'].value_counts())
```

```python
df['os']=df['10 What is the operating system (OS) of your device?\t\t\t\t                                          ']
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='os',data=df)      
plt.title('no. of order made by which operating system')
plt.show()

print('1 = windows')
print('2 = android')
print('3 = mac')

print(df['os'].value_counts())
```

```python
df['payment_mode']=df['15 What is your preferred payment Option?\t\t\t\t\t                                                                                                                               ']
```

```python
plt.subplots(figsize = (6,6))
sns.countplot(x='payment_mode',data=df)      
plt.title('modes while doing payment')
plt.show()

print('1 = cards')
print('2 = cod')
print('4 = e-wallets')

print(df['payment_mode'].value_counts())
```

```python
plt.subplots(figsize = (6,6))
sns.countplot(x='28 Being able to guarantee the privacy of the customer',data=df)      
plt.title('guarantee the privacy of the customer')
plt.show()

print('3 =agree ')
print('5 =strongly agree ')
print('4 = indifferent')

print(df['28 Being able to guarantee the privacy of the customer'].value_counts())
```

```python
plt.subplots(figsize = (16,6))
sns.countplot(x='31 Enjoyment is derived from shopping online',data=df)      
plt.title('Enjoy while doing shopping online')
plt.show()

print('1 = strongly disagree')
print('2 = disagree')
print('3 = indifferent ')
print('4 = agree')
print('5 = strongly agree ')

print(df['31 Enjoyment is derived from shopping online'].value_counts())
```

```python
df['replacement']=df['33 Return and replacement policy of the e-tailer is important for purchase decision']
```

```python
plt.subplots(figsize = (8,6))
sns.countplot(x='replacement',data=df)      
plt.title('Ease of replacing order')
plt.show()


print('2 = disagree')

print('4 = agree')
print('5 = strongly agree ')

print(df['replacement'].value_counts())
```

```python
more than 95% customers find easy to replace order.
```

```python
plt.subplots(figsize = (10,6))
sns.countplot(x='41 Monetary savings',data=df)      
plt.title('Saving while doing online shoping ')
plt.show()


print('2 = disagree')
print('3 = indifferent ')
print('4 = agree')
print('5 = strongly agree ')

print(df['41 Monetary savings'].value_counts())
```

```python
plt.subplots(figsize = (24,6))
sns.countplot(x='Which of the Indian online retailer would you recommend to a friend?',data=df)      
plt.title('recommendtion')
plt.show()

print(df['Which of the Indian online retailer would you recommend to a friend?'].value_counts())
```
